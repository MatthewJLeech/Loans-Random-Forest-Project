```python
#Importing the libraries needed 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Creating a DataFrame from a CSV file
loans = pd.read_csv('loan_data.csv')
```

```python
#Getting basic info on the DataFrame
loans.info()
```

```python
#Looking at the first couple of rows in our DataFrame
loans.head()
```

```python
#Getting some simple statistics for the numerical data in the DataFrame
loans.describe()
```

```python
#Creating a histogram showing the FICO scores with the credit policy (1 means you meet criteria)
plt.figure(figsize=(10,6))

loans[loans['credit.policy']==1]['fico'].hist(bins=35,color='blue',label='Credit Policy = 1',alpha=0.6)
loans[loans['credit.policy']==0]['fico'].hist(bins=35,color='red',label='Credit Policy = 0',alpha=0.6)

plt.legend()
plt.xlabel('FICO')
```

```python
#Creating a histogram showing the FICO scores and if they have fully paid
plt.figure(figsize=(10,6))

loans[loans['not.fully.paid']==1]['fico'].hist(bins=35,color='blue',label='Not Fully Paid = 1',alpha=0.6)
loans[loans['not.fully.paid']==0]['fico'].hist(bins=35,color='red',label='Not Fully Paid = 0',alpha=0.6)

plt.legend()
plt.xlabel('FICO')
```

```python
#Showing the reason for people wanting a loan and differenciating between if they have repaid their loan
plt.figure(figsize=(11,7))
sns.countplot(x='purpose',hue='not.fully.paid',data=loans,palette = 'Set1')
```

```python
#Showing relationship between interest rate and credit
sns.jointplot(x='fico',y='int.rate',data=loans,color='purple')
```

```python
#Graph showing the interest rate decrease with FICO scores differentiated by the credit polics and if they have repaid
plt.figure(figsize=(11,7))
sns.lmplot(y='int.rate',x='fico',data=loans,hue='credit.policy',
           col='not.fully.paid',palette='Set1')
```

```python
loans.info()
```

```python
cat_feats = ['purpose']
```

```python
#Creating a final DataFrame to change categorical columns to boolean values
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
final_data.info()
```

```python
final_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Splitting final DataFrame into inputs and outputs, then partitioning that into training and testing data
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#Importing the machine learning model
from sklearn.tree import DecisionTreeClassifier
```

```python
#Instanciating model
dtree = DecisionTreeClassifier()
```

```python
#Training model
dtree.fit(X_train,y_train)
```

```python
#Getting predictions
predictions = dtree.predict(X_test)
```

```python
#Getting metrics
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
#Now using random forest model
from sklearn.ensemble import RandomForestClassifier
```

```python
#Instanciating model
rfc = RandomForestClassifier(n_estimators=300)
```

```python
#Training the model
rfc.fit(X_train,y_train)
```

```python
#Using testing data in the trained model
predictions = rfc.predict(X_test)
```

```python
#Running the results
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
